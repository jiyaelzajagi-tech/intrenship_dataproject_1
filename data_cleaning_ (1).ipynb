```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('ecommerce_product_dataset2.csv')

df.head()
```

```python

total_duplicates = df.duplicated().sum()
print(f"Total duplicate rows: {total_duplicates}")
```

```python

df.describe()
```

which categories sell the most

```python
plt.figure(figsize=(10, 8))

sns.barplot(x='Sales', y='Category', data=df)

plt.title('Average Sales per Product Category')

plt.show()
```

Finding out which specific items sold the most

```python

plt.figure(figsize=(10, 6))

sns.barplot(x='Sales', y='Category', data=df)

plt.title('Average Sales per Product Category')
plt.xlabel('Average Sales')
plt.ylabel('Category')

plt.show()
```
